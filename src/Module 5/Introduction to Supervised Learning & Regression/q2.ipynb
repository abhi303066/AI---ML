
Question 2: Polynomial Regression <br>

Description:<br>
Perform polynomial regression to fit a non-linear dataset.

```python
# 1. Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# 2. Create a non-linear dataset
# Example: y = x^2 + noise
np.random.seed(0)
X = np.linspace(-3, 3, 100).reshape(-1, 1)
y = X**2 + np.random.randn(100, 1) * 1.5

# 3. Transform features to polynomial (degree = 2)
poly_features = PolynomialFeatures(degree=2)
X_poly = poly_features.fit_transform(X)

# 4. Train the model
model = LinearRegression()
model.fit(X_poly, y)
y_pred = model.predict(X_poly)

# 5. Evaluate the model
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"MSE: {mse:.2f}")
print(f"R² Score: {r2:.2f}")

# 6. Plot the results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Polynomial Fit (Degree 2)')
plt.title("Polynomial Regression")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
