**Task**: For the given datasets below, identify the data types and dimensions.<br>

Task 1: Employee Dataset<br>

Task 2: Product Sales Dataset<br>

Instructions:<br>

Identify which columns are numerical (continuous or discrete) and which are categorical
(nominal or ordinal).<br>
Note down the dimensions (number of rows and columns) of the dataset.

```python
import pandas as pd

# Task 1: Employee Dataset
employee_data = {
    'Employee ID': ['E001', 'E002', 'E003'],
    'Name': ['John Doe', 'Jane Smith', 'Sam Brown'],
    'Age': [34, 29, 45],
    'Department': ['HR', 'Engineering', 'Marketing'],
    'Salary': [55000, 75000, 65000],
    'Joining Date': ['2015-06-01', '2018-02-15', '2012-08-10']
}
employee_df = pd.DataFrame(employee_data)

# Task 2: Product Sales Dataset
product_sales_data = {
    'Product ID': ['P001', 'P002', 'P003'],
    'Product Name': ['Laptop', 'Phone', 'Tablet'],
    'Price': [1000, 500, 300],
    'Quantity Sold': [200, 150, 300],
    'Category': ['Electronics', 'Electronics', 'Electronics'],
    'Sales Date': ['2022-01-15', '2022-02-18', '2022-03-22']
}
product_sales_df = pd.DataFrame(product_sales_data)

# Task 3: Student Grades Dataset
student_grades_data = {
    'Student ID': ['S001', 'S002', 'S003'],
    'Student Name': ['Alice', 'Bob', 'Charlie'],
    'Math Score': [88, 92, 79],
    'Science Score': [85, 91, 82],
    'English Score': [90, 87, 85],
    'Year': [2, 3, 2]
}
student_grades_df = pd.DataFrame(student_grades_data)

# Function to analyze the dataset
def analyze_dataset(df):
    print(f"Dimensions: {df.shape}")
    print("Data Types:")
    print(df.dtypes)
    print("\nCategorical Columns:")
    categorical_cols = df.select_dtypes(include=['object']).columns
    print(categorical_cols)
    print("\nNumerical Columns:")
    numerical_cols = df.select_dtypes(include=['number']).columns
    print(numerical_cols)
    print("\n" + "="*50)

# Analyze Employee Dataset
print("Task 1: Employee Dataset")
analyze_dataset(employee_df)

# Analyze Product Sales Dataset
print("Task 2: Product Sales Dataset")
analyze_dataset(product_sales_df)

# Analyze Student Grades Dataset
print("Task 3: Student Grades Dataset")
analyze_dataset(student_grades_df)
```

**Title**: Handling Outliers<br>

**Task**: Detect and propose handling methods for outliers in the numerical columns of the datasets.<br>

Task 1: Age in Employee Dataset<br>
Task 2: Price in Product Sales Dataset<br>
Task 3: Math Score in Student Grades Dataset<br>

Instructions:<br>

Use box plots to visualize potential outliers.<br>
Suggest methods to handle them, such as removal or transformation.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Task 1: Employee Dataset
employee_data = {
    'Employee ID': ['E001', 'E002', 'E003'],
    'Name': ['John Doe', 'Jane Smith', 'Sam Brown'],
    'Age': [34, 29, 45],
    'Department': ['HR', 'Engineering', 'Marketing'],
    'Salary': [55000, 75000, 65000],
    'Joining Date': ['2015-06-01', '2018-02-15', '2012-08-10']
}
employee_df = pd.DataFrame(employee_data)

# Task 2: Product Sales Dataset
product_sales_data = {
    'Product ID': ['P001', 'P002', 'P003'],
    'Product Name': ['Laptop', 'Phone', 'Tablet'],
    'Price': [1000, 500, 300],
    'Quantity Sold': [200, 150, 300],
    'Category': ['Electronics', 'Electronics', 'Electronics'],
    'Sales Date': ['2022-01-15', '2022-02-18', '2022-03-22']
}
product_sales_df = pd.DataFrame(product_sales_data)

# Task 3: Student Grades Dataset
student_grades_data = {
    'Student ID': ['S001', 'S002', 'S003'],
    'Student Name': ['Alice', 'Bob', 'Charlie'],
    'Math Score': [88, 92, 79],
    'Science Score': [85, 91, 82],
    'English Score': [90, 87, 85],
    'Year': [2, 3, 2]
}
student_grades_df = pd.DataFrame(student_grades_data)

# Function to plot boxplot and analyze outliers
def plot_and_handle_outliers(df, column, dataset_name):
    print(f"Boxplot for {column} in {dataset_name}:")
    sns.boxplot(x=df[column])
    plt.title(f"Boxplot of {column} in {dataset_name}")
    plt.show()

    # Calculate Q1, Q3, and IQR to detect outliers
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identifying outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    
    if not outliers.empty:
        print(f"Outliers detected in {column}:\n", outliers)
        print(f"Suggested Handling Method: Removal or Transformation (e.g., log transformation for skewed data).")
    else:
        print(f"No outliers detected in {column}.")
    print("="*50)

# Task 1: Detect outliers in "Age" column (Employee Dataset)
plot_and_handle_outliers(employee_df, "Age", "Employee Dataset")

# Task 2: Detect outliers in "Price" column (Product Sales Dataset)
plot_and_handle_outliers(product_sales_df, "Price", "Product Sales Dataset")

# Task 3: Detect outliers in "Math Score" column (Student Grades Dataset)
plot_and_handle_outliers(student_grades_df, "Math Score", "Student Grades Dataset")
```

**Title**: Finding Relationships Between Features<br>

**Task**: Identify relationships between pairs of features in the datasets.<br>

Task 1: Salary vs Age in Employee Dataset<br>
Task 2: Price vs Quantity Sold in Product Sales Dataset<br>
Task 3: Math Score vs Science Score in Student Grades Dataset

Instructions: <br>

Use scatter plots or correlation coefficients to analyze the relationships.<br>
Describe any insights or patterns observed.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Task 1: Employee Dataset
employee_data = {
    'Employee ID': ['E001', 'E002', 'E003'],
    'Name': ['John Doe', 'Jane Smith', 'Sam Brown'],
    'Age': [34, 29, 45],
    'Department': ['HR', 'Engineering', 'Marketing'],
    'Salary': [55000, 75000, 65000],
    'Joining Date': ['2015-06-01', '2018-02-15', '2012-08-10']
}
employee_df = pd.DataFrame(employee_data)

# Task 2: Product Sales Dataset
product_sales_data = {
    'Product ID': ['P001', 'P002', 'P003'],
    'Product Name': ['Laptop', 'Phone', 'Tablet'],
    'Price': [1000, 500, 300],
    'Quantity Sold': [200, 150, 300],
    'Category': ['Electronics', 'Electronics', 'Electronics'],
    'Sales Date': ['2022-01-15', '2022-02-18', '2022-03-22']
}
product_sales_df = pd.DataFrame(product_sales_data)

# Task 3: Student Grades Dataset
student_grades_data = {
    'Student ID': ['S001', 'S002', 'S003'],
    'Student Name': ['Alice', 'Bob', 'Charlie'],
    'Math Score': [88, 92, 79],
    'Science Score': [85, 91, 82],
    'English Score': [90, 87, 85],
    'Year': [2, 3, 2]
}
student_grades_df = pd.DataFrame(student_grades_data)

# Task 1: Salary vs Age in Employee Dataset
plt.figure(figsize=(8, 6))
sns.scatterplot(x=employee_df['Age'], y=employee_df['Salary'])
plt.title('Salary vs Age in Employee Dataset')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()

# Compute correlation coefficient for Salary vs Age
salary_age_corr = employee_df['Age'].corr(employee_df['Salary'])
print(f"Correlation coefficient between Salary and Age: {salary_age_corr}")

# Task 2: Price vs Quantity Sold in Product Sales Dataset
plt.figure(figsize=(8, 6))
sns.scatterplot(x=product_sales_df['Price'], y=product_sales_df['Quantity Sold'])
plt.title('Price vs Quantity Sold in Product Sales Dataset')
plt.xlabel('Price')
plt.ylabel('Quantity Sold')
plt.show()

# Compute correlation coefficient for Price vs Quantity Sold
price_quantity_corr = product_sales_df['Price'].corr(product_sales_df['Quantity Sold'])
print(f"Correlation coefficient between Price and Quantity Sold: {price_quantity_corr}")

# Task 3: Math Score vs Science Score in Student Grades Dataset
plt.figure(figsize=(8, 6))
sns.scatterplot(x=student_grades_df['Math Score'], y=student_grades_df['Science Score'])
plt.title('Math Score vs Science Score in Student Grades Dataset')
plt.xlabel('Math Score')
plt.ylabel('Science Score')
plt.show()

# Compute correlation coefficient for Math Score vs Science Score
math_science_corr = student_grades_df['Math Score'].corr(student_grades_df['Science Score'])
print(f"Correlation coefficient between Math Score and Science Score: {math_science_corr}")
```
