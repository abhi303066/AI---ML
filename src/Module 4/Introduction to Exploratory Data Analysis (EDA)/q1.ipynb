Question 4: Handling Outliers
<br>
Description: Identify outliers in numerical columns using box plots.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Plot boxplots for each numerical feature
plt.figure(figsize=(12, 8))
for i, column in enumerate(df.columns, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(y=df[column], color='lightblue')
    plt.title(f'Boxplot of {column}')
    plt.tight_layout()

plt.show()
```


Question 5: Categorical Data Analysis
<br>
Description: Explore the counts of categorical variables.

```python
import pandas as pd
from sklearn.datasets import load_iris

# Load dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['target'] = iris.target

# Map numeric target to species names for better readability
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Count the occurrences of each species
print("Count of each species (categorical variable):")
print(df['species'].value_counts())

```

Question 6: Data Transformation
<br>
Description: Transform a categorical column into numerical using Label Encoding.

```python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.preprocessing import LabelEncoder

# Load the dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add species as a categorical column
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

# Apply Label Encoding
label_encoder = LabelEncoder()
df['species_encoded'] = label_encoder.fit_transform(df['species'])

# Show mapping and result
print("Label Encoding Mapping:")
for i, class_ in enumerate(label_encoder.classes_):
    print(f"{class_}: {i}")

print("\nFirst few rows with encoded species:")
print(df[['species', 'species_encoded']].head())

```

Question 8: Correlation Analysis
<br>
Description: Calculate and visualize the correlation matrix for numerical features.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

# Load the dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlation Matrix of Numerical Features", fontsize=16)
plt.show()
```

Question 10: Advanced Outlier Detection
<br>
Description: Use the Z-score method to identify and handle outliers.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from scipy.stats import zscore

# Load the Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Calculate the Z-scores for each numerical column
z_scores = np.abs(zscore(df))

# Set a threshold for identifying outliers (commonly 3)
threshold = 3

# Identify the outliers (rows where any column has a Z-score greater than threshold)
outliers = (z_scores > threshold).any(axis=1)

# Display the rows that are outliers
outlier_rows = df[outliers]

# Display the outliers
print("Outliers detected:")
print(outlier_rows)
```
