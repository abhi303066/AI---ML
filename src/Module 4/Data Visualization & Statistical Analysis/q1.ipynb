Title: Histogram of a Single Feature <br>

Question 1: Create a histogram for the age feature from a dataset. Interpret what the shape of the histogram tells us about the distribution of the age feature.

Title: Correlation Heatmap<br>

Question 5 : Generate a correlation heatmap for a dataset with multiple features (e.g., height ,weight , age ) and explain the correlations observed.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset with height, weight, and age columns (replace with your actual dataset)
data = {'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan'],
        'Height': [150, 160, 175, 165, 180],
        'Weight': [55, 65, 80, 70, 85],
        'Age': [25, 30, 35, 28, 40]}

df = pd.DataFrame(data)

# Compute the correlation matrix
correlation_matrix = df[['Height', 'Weight', 'Age']].corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)

# Add title
plt.title('Correlation Heatmap for Height, Weight, and Age')

# Show the plot
plt.show()
```

Title: Descriptive Statistical Analysis<br>

Question 6: Calculate the mean, median, standard deviation, skewness, and kurtosis for the
temperature feature and discuss the results.

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

# Example dataset with temperature column (replace with your actual dataset)
data = {'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix'],
        'Temperature': [15.5, 22.3, 10.8, 27.2, 33.1]}  # Temperature in Celsius

df = pd.DataFrame(data)

# Calculate mean, median, standard deviation, skewness, and kurtosis for the 'Temperature' feature
mean_temp = df['Temperature'].mean()
median_temp = df['Temperature'].median()
std_temp = df['Temperature'].std()
skewness_temp = skew(df['Temperature'])
kurtosis_temp = kurtosis(df['Temperature'])

# Display the results
print(f"Mean: {mean_temp:.2f}")
print(f"Median: {median_temp:.2f}")
print(f"Standard Deviation: {std_temp:.2f}")
print(f"Skewness: {skewness_temp:.2f}")
print(f"Kurtosis: {kurtosis_temp:.2f}")
```

Title: Covariance and Correlation between Two Features<br>

Question 7: Compute the covariance and correlation between price and demand in a dataset.
Explain what these metrics indicate.

```python
import pandas as pd

# Example dataset with price and demand columns (replace with your actual dataset)
data = {'Product': ['A', 'B', 'C', 'D', 'E'],
        'Price': [100, 150, 200, 250, 300],
        'Demand': [1200, 1100, 900, 600, 300]}

df = pd.DataFrame(data)

# Calculate covariance and correlation between 'Price' and 'Demand'
covariance = df['Price'].cov(df['Demand'])
correlation = df['Price'].corr(df['Demand'])

# Display the results
print(f"Covariance between Price and Demand: {covariance:.2f}")
print(f"Correlation between Price and Demand: {correlation:.2f}")
```

Title: Principal Component Analysis (PCA)<br>

Question 9 : Apply PCA on a dataset with multiple features (e.g., x1 , x2 , x3 , x4 ) and reduce it to two principal components. Visualize the data in the new feature space.

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Example dataset with 4 features (replace with your actual dataset)
data = {'x1': [2, 3, 4, 5, 6],
        'x2': [8, 7, 6, 5, 4],
        'x3': [1, 2, 3, 4, 5],
        'x4': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)

# Step 1: Standardize the data (mean = 0, variance = 1)
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 2: Apply PCA to reduce data to two principal components
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)

# Step 3: Visualize the data in the new 2D feature space
plt.figure(figsize=(8, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], color='blue', marker='o')

# Add labels and title
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA: Reduced to 2 Principal Components')

# Show the plot
plt.show()

# Print explained variance to understand how much information is retained
print(f"Explained Variance Ratios: {pca.explained_variance_ratio_}")
print(f"Total Variance Explained: {sum(pca.explained_variance_ratio_):.2f}")
```

Title: Advanced Pair Plot with Hue Parameter<br>

Question 10 : Create a pair plot for height , weight , and age with an added categorical variable gender as the hue to observe different group trends.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Example dataset with height, weight, age, and gender (replace with your actual dataset)
data = {'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Ethan'],
        'Height': [150, 160, 175, 165, 180],
        'Weight': [55, 65, 80, 70, 85],
        'Age': [25, 30, 35, 28, 40],
        'Gender': ['Female', 'Male', 'Male', 'Female', 'Male']}

df = pd.DataFrame(data)

# Create a pair plot with hue as 'Gender' to observe trends by gender
sns.pairplot(df[['Height', 'Weight', 'Age', 'Gender']], hue='Gender')

# Add title
plt.suptitle('Advanced Pair Plot: Height, Weight, Age by Gender', y=1.02)

# Show the plot
plt.show()
```
