Question 6: Replace missing values in a column with the median value of that column.

```python
import pandas as pd

# Load dataset from GitHub
url = 'https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv'
df = pd.read_csv(url)

# Replace missing values in 'math score' with the median of that column
median_value = df['math score'].median()
df['math score'].fillna(median_value, inplace=True)

# Verify no missing values remain in that column
print(df['math score'].isnull().sum())
```

Question 8: Create a histogram to visualize the distribution of a numerical column (e.g., 'price').

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset from GitHub
url = 'https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv'
df = pd.read_csv(url)

# Create a histogram for the 'math score' column
plt.hist(df['math score'], bins=10, color='skyblue', edgecolor='black')

# Add title and labels
plt.title('Distribution of Math Scores')
plt.xlabel('Math Score')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

Question 9: Create a scatter plot to visualize the relationship between two numerical variables

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset from GitHub
url = 'https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv'
df = pd.read_csv(url)

# Create a scatter plot to visualize the relationship between 'math score' and 'reading score'
plt.scatter(df['math score'], df['reading score'], color='purple')

# Add title and labels
plt.title('Scatter Plot: Math Score vs Reading Score')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')

# Show the plot
plt.show()
```

Question 10: Visualize the correlation matrix of the dataset using a heatmap.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset from GitHub
url = 'https://raw.githubusercontent.com/rashida048/Datasets/master/StudentsPerformance.csv'
df = pd.read_csv(url)

# Select only numerical columns
numerical_df = df.select_dtypes(include=['int64', 'float64'])

# Compute the correlation matrix
correlation_matrix = numerical_df.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```
